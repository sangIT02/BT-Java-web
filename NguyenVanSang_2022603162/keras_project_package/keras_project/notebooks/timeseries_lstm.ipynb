```
import pandas as pd, numpy as np, matplotlib.pyplot as plt, tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# đọc từ Drive nếu có, ngược lại đọc local
drive_path = '/content/drive/MyDrive/keras_project/data/timeseries.csv'
path = drive_path if tf.io.gfile.exists(drive_path) else '../data/timeseries.csv'
df = pd.read_csv(path)
values = df['value'].values.astype('float32')

# windowing
win = 30
X, y = [], []
for i in range(len(values)-win):
    X.append(values[i:i+win])
    y.append(values[i+win])
X = np.array(X)[..., None]
y = np.array(y)

split = int(len(X)*0.8)
X_train, X_val = X[:split], X[split:]
y_train, y_val = y[:split], y[split:]

model = keras.Sequential([layers.Input((win,1)), layers.LSTM(64), layers.Dense(1)])
model.compile(optimizer='adam', loss='mse')
hist = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, batch_size=64)

plt.figure(); plt.plot(hist.history['loss'], label='loss'); plt.plot(hist.history['val_loss'], label='val_loss'); plt.legend(); plt.title('MSE'); plt.xlabel('epoch'); plt.show()
pred = model.predict(X_val[:200]).squeeze()
plt.figure(); plt.plot(y_val[:200], label='true'); plt.plot(pred, label='pred'); plt.legend(); plt.title('Forecast sample'); plt.show()
```
