```
import tensorflow as tf, pandas as pd, numpy as np
from tensorflow import keras
from tensorflow.keras import layers

drive_path = '/content/drive/MyDrive/keras_project/data/nlp_texts.csv'
path = drive_path if tf.io.gfile.exists(drive_path) else '../data/nlp_texts.csv'
df = pd.read_csv(path)
texts = df['text'].astype(str).tolist(); labels = df['label'].values.astype('float32')

# Fallback: TextVectorization (ổn định trên Colab)
vec = layers.TextVectorization(output_sequence_length=40, max_tokens=5000)
vec.adapt(tf.data.Dataset.from_tensor_slices(texts).batch(32))

X = vec(np.array([[t] for t in texts])); y = labels
model = keras.Sequential([layers.Embedding(5000, 64), layers.GlobalAveragePooling1D(), layers.Dense(1, activation='sigmoid')])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=6, batch_size=16, validation_split=0.2)
```
