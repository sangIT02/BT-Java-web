```
import numpy as np, matplotlib.pyplot as plt, tensorflow as tf
import keras_cv
from PIL import Image

img_path = '/content/drive/MyDrive/keras_project/data/images/objects_sample.png'
if not tf.io.gfile.exists(img_path):
    img_path = '../data/images/objects_sample.png'
img = np.array(Image.open(img_path).convert('RGB'))

detector = keras_cv.models.RetinaNet.from_preset('retinanet_resnet50_v1_fpn_coco')
pred = detector.predict(np.expand_dims(img, 0))

boxes = pred['boxes'][0].numpy()
scores = pred['confidence'][0].numpy()

plt.figure(figsize=(6,4)); plt.imshow(img)
for b, s in zip(boxes[:10], scores[:10]):
    if s < 0.25: continue
    y1, x1, y2, x2 = b
    plt.gca().add_patch(plt.Rectangle((x1,y1), x2-x1, y2-y1, fill=False, linewidth=2))
plt.axis('off'); plt.title('KerasCV RetinaNet demo'); plt.show()
```
